```python
# Convert runs and error_rates to numpy arrays
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the provided CSV file
file_path = 'summary.csv'
data = pd.read_csv(file_path)
data['error_rate'] = round((data['errors'] / (data['correct'] + data['errors'])) * 100, 1)

runs = np.array(data['run'])
error_rates = np.array(data['error_rate'])

# Create the plot
plt.figure(figsize=(12, 8))
plt.plot(runs, error_rates, marker='o', linestyle='-', color='orange', label='Error Rate')

average_error_rate = np.mean(error_rates)
plt.axhline(y=average_error_rate, color='darkgrey', linestyle=':', linewidth=2, label=f'Average Error Rate: {average_error_rate:.1f}%')

# Set plot title and labels
plt.title('Analysis of Errors', fontsize=18)
plt.xlabel('Experiment', fontsize=15)
plt.ylabel('Error Rate (%)', fontsize=15)

# Customize plot aesthetics to match PostNL colors
plt.grid(True, which='both', linestyle='--', linewidth=0.7, color='grey')
plt.axhline(y=0, color='grey', linewidth=1)
plt.axvline(x=0, color='grey', linewidth=1)
plt.fill_between(runs, error_rates, color='#fca47c', alpha=0.1)

# Set the plot background to white
plt.gca().set_facecolor('white')

# Show all x-axis labels
plt.xticks(runs, fontsize=12, rotation=45)

# Show all y-axis labels
plt.yticks(fontsize=12)

# Display legend
plt.legend(fontsize=12)

# Save the plot as a PNG file
plt.savefig('experiments.png', bbox_inches='tight', dpi=300)

# Show the plot
plt.show()
```

```python
import seaborn as sns

mean_error_rate = data['error_rate'].mean()
std_error_rate = data['error_rate'].std()

# Define a color similar to the PostNL logo orange
postnl_orange = '#FF6600'  # Hex code for a typical bright orange color

# Create a styled box plot using the specified orange color
plt.figure(figsize=(10, 6))
sns.boxplot(y='error_rate', data=data[['run','error_rate']], color=postnl_orange)
plt.title('Experiment Error Rates (N=17)', fontsize=16, fontweight='bold')
plt.ylabel('Error Rate (%)', fontsize=14)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.legend([f'Mean: {mean_error_rate:.2f}', f'Standard Deviation: {std_error_rate:.2f}'], loc='upper left')

plt.savefig('boxplot.png', bbox_inches='tight', dpi=600)

# Display the plot
plt.show()
```
